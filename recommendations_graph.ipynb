```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
recs_df = pd.read_csv('recommendations.csv')
recs_df
```

```python
recs_format = recs_df.drop(columns=['target_id', 'source_id'])
recs_format = recs_format[['source_title', 'target_title', 'number']]
recs_format
```

```python
# nxG = nx.Graph()
# nxG.add_node(1)
# nxG.add_node(2)
nxG = nx.from_pandas_edgelist(recs_format, source='source_title', target='target_title', edge_attr='number')
```

```python
nx.draw(nxG)
plt.savefig('network_graph/graph.png')
```
